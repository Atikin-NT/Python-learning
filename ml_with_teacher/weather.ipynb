```python
import numpy as np
import pandas as pd

df = pd.read_csv('./weather.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df['RainToday']=df['RainToday'].apply(lambda x:1 if x == "Yes" else 0)
df['RainTomorrow']=df['RainTomorrow'].apply(lambda x:1 if x == "Yes" else 0)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df=df.dropna()
df.shape
```

```python
df.WindGustDir = le.fit_transform(df.WindGustDir)
df.WindDir3pm = le.fit_transform(df.WindDir3pm)
df.WindDir9am = le.fit_transform(df.WindDir9am)
df.describe()
```

```python
x = df.drop(columns=['Date', 'RainTomorrow'])
y = df['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split

train_x, train_y, test_x, test_y = train_test_split(x, y, test_size=0.2, random_state=2)
```

Поскольку метка является двоичной переменной, то мы обучаем классификатор. Для нашей задачи лучше всего подходит логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(train_x, test_x)
```

```python
predict = classifier.predict(train_y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(predict, test_y)
```
